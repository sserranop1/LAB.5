```python
from pathlib import Path

carpeta = Path("/home/santiagos/Desktop/Universidad/intersemestral/LAB.5/")
salida = Path("graficas")
salida.mkdir(exist_ok=True)

imagenes = ["m92010", "m92011", "m92014", "m92015"]
aperturas = [1, 2, 3, 4, 5, 10]

def leer_archivo(ruta):
    df = pd.read_csv(
        ruta,
        sep=r'\s+',
        names=["XCENTER", "YCENTER", "MAG", "MERR"],
        na_values=["INDEF"]
    )
    df = df.dropna()
    df = df[(df["MERR"] > 0) & (df["MERR"] < 1)]
    return df

resumen = []

for imagen in imagenes:
    plt.figure(figsize=(8, 6))

    for ap in aperturas:
        archivo = carpeta / f"{imagen}_ap{ap}F.dat"

        if not archivo.exists():
            print(f"No existe {archivo}")
            continue

        df = leer_archivo(archivo)

        resumen.append({
            "imagen": imagen,
            "apertura_FWHM": ap,
            "n_estrellas": len(df),
            "merr_mediana": df["MERR"].median(),
            "merr_promedio": df["MERR"].mean(),
            "merr_p75": df["MERR"].quantile(0.75)
        })

        plt.scatter(
            df["MAG"],
            df["MERR"],
            s=8,
            alpha=0.6,
            label=f"{ap}*FWHM"
        )

    plt.xlabel("Magnitud instrumental MAG")
    plt.ylabel("Error fotométrico MERR")
    plt.title(f"MAG vs MERR para diferentes aperturas - {imagen}")
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.ylim(bottom=0)
    plt.savefig(salida / f"{imagen}_aperturas_mag_vs_merr.png", dpi=200)
    plt.close()

resumen = pd.DataFrame(resumen)
resumen.to_csv(salida / "resumen_aperturas.csv", index=False)

print(resumen)
```

```python
carpeta = Path("/home/santiagos/Desktop/Universidad/intersemestral/LAB.5/")

imagenes = ["m92010", "m92011", "m92014", "m92015"]
thresholds = [3, 4, 5, 6, 7, 8, 10]

filas = []

for imagen in imagenes:
    for th in thresholds:
        archivo = carpeta / f"{imagen}_thr{th}.xy"

        if not archivo.exists():
            print(f"No existe {archivo}")
            continue

        df = pd.read_csv(
            archivo,
            sep=r'\s+',
            names=["XCENTER", "YCENTER", "ID"],
            na_values=["INDEF"]
        )

        filas.append({
            "imagen": imagen,
            "threshold": th,
            "n_detectadas": len(df),
            "n_falsas": "",
            "n_reales": "",
            "n_perdidas": "",
            "comentario": ""
        })

tabla = pd.DataFrame(filas)
tabla.to_csv("graficas/tabla_thresholds_para_llenar.csv", index=False)

print(tabla)
```

```python

carpeta = Path("/home/santiagos/Desktop/Universidad/intersemestral/LAB.5/")
salida = Path("graficas")
salida.mkdir(exist_ok=True)

imagenes = ["m92010", "m92011", "m92014", "m92015"]

resumen = []

for imagen in imagenes:
    archivo = carpeta / f"{imagen}_final.dat"

    if not archivo.exists():
        print(f"No existe {archivo}")
        continue

    df = pd.read_csv(
        archivo,
        sep=r'\s+',
        names=["XCENTER", "YCENTER", "MAG", "MERR"],
        na_values=["INDEF"]
    )

    df = df.dropna()
    df = df[(df["MERR"] > 0) & (df["MERR"] < 1)]

    resumen.append({
        "imagen": imagen,
        "n_estrellas_con_fotometria": len(df),
        "merr_mediana": df["MERR"].median(),
        "merr_promedio": df["MERR"].mean(),
        "mag_min": df["MAG"].min(),
        "mag_max": df["MAG"].max()
    })

    plt.figure(figsize=(8, 6))
    plt.scatter(df["MAG"], df["MERR"], s=8, alpha=0.7)
    plt.xlabel("Magnitud instrumental MAG")
    plt.ylabel("Error fotométrico MERR")
    plt.title(f"MAG vs MERR final - M41")
    plt.grid(True, alpha=0.3)
    plt.ylim(bottom=0)
    plt.savefig(salida / f"m91_final_mag_vs_merr.png", dpi=200)
    plt.close()

resumen = pd.DataFrame(resumen)
resumen.to_csv(salida / "tabla_numero_estrellas_final.csv", index=False)

print(resumen)
```
